## Task 1. 
Scrap Amazon to get data for their best seller books. Get title, author, rating, number of users rated and price information for all available pages of products. (e.g. if there are 10 pages of best seller books, you need to scrap all the pages, not just the first page). Create a pandas dataframe and keep the scrapped information inside the dataframe, in separate columns.

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
from fuzzywuzzy import fuzz
````

````python
# pip install fuzzywuzzy
````

````python
no_pages = 2

def get_data(pageNo):  
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    r = requests.get('https://www.amazon.com/best-sellers-books-Amazon/zgbs/books/ref=zg_bs_pg_'+str(pageNo)+'?_encoding=UTF8&pg='+str(pageNo), headers=headers)#, proxies=proxies)
    content = r.content
    soup = BeautifulSoup(content)
    #print(soup)

    alls = []
    for d in soup.findAll('div', attrs={'class':'a-section a-spacing-none aok-relative'}):
        #print(d)
        name = d.find('span', attrs={'class':'zg-text-center-align'})
        n = name.find_all('img', alt=True)
        #print(n[0]['alt'])
        author = d.find('a', attrs={'class':'a-size-small a-link-child'})
        rating = d.find('span', attrs={'class':'a-icon-alt'})
        users_rated = d.find('a', attrs={'class':'a-size-small a-link-normal'})
        price = d.find('span', attrs={'class':'p13n-sc-price'})

        all1=[]

        if name is not None:
            #print(n[0]['alt'])
            all1.append(n[0]['alt'])
        else:
            all1.append("unknown-product")

        if author is not None:
            #print(author.text)
            all1.append(author.text)
        elif author is None:
            author = d.find('span', attrs={'class':'a-size-small a-color-base'})
            if author is not None:
                all1.append(author.text)
            else:    
                all1.append('0')

        if rating is not None:
            #print(rating.text)
            all1.append(rating.text)
        else:
            all1.append('-1')
            
        if users_rated is not None:
            #print(price.text)
            all1.append(users_rated.text)
        else:
            all1.append('0')     

        if price is not None:
            #print(price.text)
            all1.append(price.text)
        else:
            all1.append('0')
        alls.append(all1)    
    return alls
````

````python
results = []
for i in range(no_pages+1):
    results.append(get_data(i))
flatten = lambda l: [item for sublist in l for item in sublist]
df = pd.DataFrame(flatten(results),columns=['Book Name','Author','Rating','Customers_Rated', 'Price'])
# df.to_csv('C:/Users/ASUS/Downloads/amazon_books.csv', index=False, encoding='utf-8')
````

````python
class Book:
    
    def __init__(self, title, author, rating, num_users_rated, price):
        """
        Create the attributes title, author, rating, num_users_rated and price
        
        """
        self.title = title
        self.author = author
        self.rating = rating
        self.num_users_rated = num_users_rated
        self.price = price
        
        pass
    
    def pretty_print(self):
        """
        Print book information in the following format:
        $title by $author with rating $rating costs $price.
        
        """
        print("%s by %s with rating %s costs %s" % (self.title, self.author, self.rating, self.price))
        
        pass
    
    def rating_eval(self):
        """
        Evaluates the statistical significance of the rating. The method should return 
        "statistically significant rating" (or ssr) in case more than 5000 users have
        rated the following book and should return "not statistically significant rating" (or nssr), otherwise.
        
        """
        if self.num_users_rated > 5000:
            print('statistically significant rating')
        else: 
            print('not statistically significant rating')
        
        pass
````

````python
class Library:
    
    def __init__(self, Book_list):
        """
        Create an attribute of type list called Book_list which will be an empty list for now.

        """
        self.Book_list = Book_list
#        print(self.Book_list)
        pass
    
    def get_top_5(self):
        """
        Return the information related to 5 books that have the highest rating using the method pretty_print.
        If there are more than 5 books with the same rating, randomly select and show only 5 of them.
        
        """
# I could create a new column extracting rating from the string 
# but I think sorting by an int column is better and almost the same 

        book = [list([self.Book_list[i][0],self.Book_list[i][1], self.Book_list[i][2], self.Book_list[i][3], self.Book_list[i][4]]) for i in range(len(self.Book_list))]
        def Sort(sub_li):
            l = len(sub_li)
            for k in range(0, l):
                for j in range(0, l-k-1):
                    if (sub_li[j][3] < sub_li[j + 1][3]):
                        tempo = sub_li[j]
                        sub_li[j]= sub_li[j + 1]
                        sub_li[j + 1]= tempo
            return sub_li
        
        for each_book in Sort(book)[:5]:
                each_book = Book(each_book[0],each_book[1],each_book[2],each_book[3],each_book[4])
                print(each_book.pretty_print())

    def simple_search(self):
        """
        Search a book with required title and print information related to all books with titles exactly 
        matching the searched title using the method pretty_print.
        
        """
        book = [list([self.Book_list[i][0],self.Book_list[i][1], self.Book_list[i][2], self.Book_list[i][3], self.Book_list[i][4]]) for i in range(len(self.Book_list))]
        while True:
            title_imp = input()
            if title_imp not in [x[0] for x in book]:
                print("No such a book in the list")
                break
            else:
#                print(Book.pretty_print(self.title == title_imp))
                for the_book in book:
                        if the_book[0] == title_imp:
                            the_book = Book(the_book[0],the_book[1],the_book[2],the_book[3],the_book[4])
                            print(the_book.pretty_print())
                            break
            break
        pass
    
    def complex_search(self):
        """
        In this case the user may not remember the exact title of the book. If you have a book(s) with 
        title exactly matching the searched title, this method should perform just like the method simple_search.
        Otherwise, if you don't have any books with the searched title, you should return book(s) that have the
        most similar titles. You are free to implement this method however you like. 
        One version that you can implement in case you don't have other ideas:
        return the book(s) with titles containing the most number of words from the searched title 
        (return the information using the method pretty_print), and if there are no such books, 
        return "nothing macthed your search".
        I would be happy to see any other interesting versions of the method :)
        
        """
        book = [list([self.Book_list[i][0],self.Book_list[i][1], self.Book_list[i][2], self.Book_list[i][3], self.Book_list[i][4]]) for i in range(len(self.Book_list))]
        while True:
            title_imp = input()
            if title_imp not in [x[0] for x in book]:
                    for example in [x[0] for x in book]:
                        if fuzz.token_set_ratio(example, title_imp) > 70:
                        # https://pythoninoffice.com/how-to-find-similar-strings-using-python/
                            for similar_book in book:
                                    if similar_book[0] == example:
                                        similar_book = Book(similar_book[0],similar_book[1],similar_book[2],similar_book[3],similar_book[4])
                                        print(similar_book.pretty_print())
                                    elif similar_book not in book:
                                        print('nothing really match the search')
            else:
#                print(Book.pretty_print(self.title == title_imp))
                for the_book in book:
                        if the_book[0] == title_imp:
                            the_book = Book(the_book[0],the_book[1],the_book[2],the_book[3],the_book[4])
                            print(the_book.pretty_print())
                            break
            

        pass
````
