## Problem 1

```
Let say you want to send a message to someone. You have 2 options - you can Email the message or SMS the message to the corresponding person. So, you have two options to send the message and the client side code will use one of the implementations to send the message to the corresponding person. Use Bridge Design Pattern to implement the logic with classes of your choice and make sure to test the implementation with some concrete objects.
```

````python
from __future__ import annotations
from abc import ABC, abstractmethod

class ProducingAPI(ABC):
 
    
    @abstractmethod
    def produceMes(self, text, who):
 
        pass

class Email(ProducingAPI):
 
    """Implementation specific"""
 
    def produceMes(self, text, who):
 
        print(f'New message from {who} - '
              f' "{text}"')
 
class SMS(ProducingAPI):
 
    """Implementation specific"""
 
    def produceMes(self, text, who):
 
        print(f'New message from {who} - '
              f' "{text}"')
 
class SendMessage:
 
    def __init__(self, text, who, producingAPI):
 
        """Implementation independent"""
 
        self._text = text
        self._who = who
 
        self._producingAPI = producingAPI
 
    def produce(self):
 
        self._producingAPI.produceMes(self._text, self._who)
 
    def expand(self, times):
 
        """Implementation independent"""
 
        self._text = self._text + "End of conv"
 
 

 
Mes1 = SendMessage("Hello","Anna", SMS())
Mes1.produce()
 
Mes2 = SendMessage("Congrats bro", "Hayk", Email())
Mes2.produce()
````

## Problem 2

```
Follow the diagram below and use Composite Design Pattern to implement and test the logic. Add any other classes of your choice. The top of your hierchy will be the box which will contain Instrument Collection which will contain instruments. 
```

````python
class Instrument:
  
    '''Class representing objects at the bottom or Leaf of the hierarchy tree.'''
  
    def __init__(self, name):
  
        ''''Takes the first positional argument and assigns to member variable "position".'''
        self.name = name
  
    def showDetails(self):
  
        '''Prints the child element name.'''
        print("\t", end ="")
        print(f"contains {self.name}")
  
  
class Box:
  
    '''Class representing objects at any level of the hierarchy
     tree except for the bottom or leaf level. Maintains the child
      objects by adding and removing them from the tree structure.'''
  
    def __init__(self, name):
  
        '''Takes the first positional argument and assigns to member
         variable "position". Initializes a list of children elements.'''
        self.name = name
        self.instruments = []
  
    def add(self, instrument):
  
        '''Adds the supplied child element to the list of children
         elements "children".'''
        self.instruments.append(instrument)
  
    def remove(self, instrument):
  
        '''Removes the supplied child element from the list of
        children elements "children".'''
        self.instruments.remove(instrument)
  
    def showDetails(self):
  
        '''Prints the details of the component element first. Then,
        iterates over each of its children, prints their details by
        calling their showDetails() method.'''
        print(self.name)
        names = self.report_box_content()
        print(names)
        for instrument in self.instruments:
            print("\t", end ="")
            instrument.showDetails()
            
    def report_box_content(self):
        return self.instruments
  

  

box = Box("Box1")
instrument1 = Instrument("Pen")
instrument2 = Instrument("Pencil")
instrument3 = Instrument("Rubber")

box.add(instrument1)
box.add(instrument2)
box.add(instrument3)

box.showDetails()
````
