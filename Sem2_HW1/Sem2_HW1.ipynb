## Problem 2
Use Factory Method pattern to create 2 different classes PersonalAccount and
BusinessAccount of type BankAccount having a method CreateAccount with some
implementation of your choice. Have a BankAccountFactory class and a Client class and allow
the Client to ask the factory to create an account. The type of the account should be determined
based on the clientâ€™s input. You can decide what should be the client input and how the factory
class should determine what type of an account to create.

```python
class BankAccount:
    def __init__(self): #receives instance 'self' as argument
        print('Our BankAccount class is created')
```

```python
class PersonalAccount(BankAccount):
    def __init__(self, legalstatus, code):
        self.legalstatus = legalstatus
        self.code = code

    def CreateAccount(self):
        return "PA" + self.code

class BusinessAccount(BankAccount):
    def __init__(self, legalstatus, code):
        self.legalstatus = legalstatus
        self.code = code

    def CreateAccount(self):
        return "BA" + self.code
```

```python
class BankAccountFactory:
    def create_account(self, legalstatus):
        if legalstatus == 'individual':
            code = input("Enter the passport code: ")
            return PersonalAccount(str(legalstatus),str(code))

        elif legalstatus == 'legal':
            code = input("Enter the passport code: ")
            return BusinessAccount(str(legalstatus),str(code))
```

```python
class Client:
    @staticmethod
    def accounts_client():
            account_factory = BankAccountFactory()
            legalstatus = input("Enter the legal status: ")

            Account = account_factory.create_account(legalstatus)

            print(f"Generated {legalstatus} client's code: {Account.CreateAccount()}")
```

```python
test = Client()
test.accounts_client()
```

## Problem 3
Use the diagram below and function logic of your choice to apply the Abstract Factory pattern.


```python
from __future__ import annotations
from abc import ABC, abstractmethod


class AbstractFactory(ABC):

    @abstractmethod
    def create_button(self):
        pass
    
    @abstractmethod
    def create_checkbox(self):
        pass
    
class WinFactory(AbstractFactory):

    def create_button(self):
        return WinButton()

    def create_checkbox(self):
        return WinCheckbox()
    
class MacFactory(AbstractFactory):

    def create_button(self):
        return MacButton()

    def create_checkbox(self):
        return MacCheckbox()
    
class WinButton(WinFactory):
    def button_func(self):
        return 'Windows button'
    def paint(self):
        return 'black'
    
class MacButton(MacFactory):
    def button_func(self):
        return 'Mac button'
    def paint(self):
        return 'gray'
    
class WinCheckbox(WinFactory):
    def checkbox_func(self):
        return 'Win checkbox'
    def paint(self):
        return 'black'
    
class MacCheckbox(MacFactory):
    def checkbox_func(self):
        return 'Mac checkbox'
    def paint(self):
        return 'gray'
    
def client_code(factory):

    button = factory.create_button()
    checkbox = factory.create_checkbox()

    print(button.button_func() + " " + button.paint())
    print(checkbox.checkbox_func() + " " + checkbox.paint())

client_code(WinFactory())
client_code(MacFactory())
```
