## Problem 1
You know that each BankAccount object has attributes id, balance and rate. Use builder pattern and any extra classes and methods that you may need to realize the creation of a BankAccount type object. Create some objects and do some operations to test your classes.

```python
from __future__ import annotations
from abc import ABC, abstractmethod, abstractproperty
from typing import Any


class Builder(ABC):
    
    @abstractmethod
    def create_i_d(self, _id):
        pass

    @abstractmethod
    def create_balance(self, _balance):
        pass
    
    @abstractmethod
    def create_rate(self, _rate):
        pass

    
    def create_socialID(self, _socialID):
        pass
    
    def create_HVHH(self, _hvhh):
        pass
    


class Business(Builder):


    def __init__(self):
        self.BankAccount = BankAccount()

    def create_i_d(self, _id):
        self.BankAccount.add(_id)

    def create_balance(self, _balance):
        self.BankAccount.add(_balance)
        
    def create_rate(self, _rate):
        self.BankAccount.add(_rate)
        
    def create_socialID(self, _socialID):
        self.BankAccount.add(_socialID)
        

class Individual(Builder):


    def __init__(self):
        self.BankAccount = BankAccount()

    def create_i_d(self, _id):
        self.BankAccount.add(_id)

    def create_balance(self, _balance):
        self.BankAccount.add(_balance)
        
    def create_rate(self, _rate):
        self.BankAccount.add(_rate)

    def create_HVHH(self, _hvhh):
        self.BankAccount.add(_hvhh)



class BankAccount():

    def __init__(self):
        self.decription = []

    def add(self, desc: int):
        self.decription.append(desc)

    def describe(self):
        print(f"Account code {', '.join(self.decription)}", end="")



builder = Business()

print("Custom account1: ")

builder.create_i_d('0001')
builder.create_balance('1000000')
builder.BankAccount.describe()
              
builder = Individual()

print("\nCustom account2: ")

builder.create_i_d('0002')
builder.create_balance('2000000')
builder.create_HVHH('0123456')
builder.BankAccount.describe()
```

## Problem 2
Use prototype design pattern and classes of your choice. create an abstract class Shape and concrete classes extending the Shape class: Circle, Square and Rectangle. Define a class ShapeCache which stores shape objects in a dictionary and returns their clones when requested. Create some objects and do some operations to test your classes.

```python
from abc import ABCMeta, abstractmethod
import copy
  
  

class Shapes(metaclass = ABCMeta):
      
    def __init__(self):
        self.type = None
        self.id = None
        self.width = None
        self.radius = None
        self.height = None
  
    @abstractmethod
    def shape(self):
        pass
  
    def get_type(self):
        return self.type
    
    def get_id(self):
        return self.id
  
    def set_height(self):
        self.height = height
  
    def set_width(self):
        self.width = width
        
    def set_radius(self):
        self.radius = radius
        
    def set_id(self, sid):
        self.id = sid    
  
    def clone(self):
        return copy.copy(self)
  

class Circle(Shapes):
    def __init__(self, color):
        self.color = color
        self.type = "Circle"
  
    def shape(self):
        return 3.14 * self.radius * self.radius
  
class Square(Shapes):
    def __init__(self, color):
        self.color = color
        self.type = "Canonical tetragon"
  
    def shape(self):
        return self.width ** 2
  
class Rectangle(Shapes):
    def __init__(self, color):
        self.color = color
        self.type = "Tetragon"
  
    def shape(self):
        return self.width * self.height
  
# cache class
class Shapes_Cache:
      
    
    cache = {}
  
    @staticmethod
    def get_shape(sid):
        SHAPE = Shapes_Cache.cache.get(sid, None)
        return SHAPE.clone() 
  
    @staticmethod
    def load():
        square = Square('red')
        square.set_id("1")
#        square.set_width(2)
        Shapes_Cache.cache[square.get_id()] = square
  
        circle = Circle('yellow')
        circle.set_id("2")
#        circle.set_radius(7)
        Shapes_Cache.cache[circle.get_id()] = circle
  
        rectangle = Rectangle('black')
        rectangle.set_id("3")
#        rectangle.set_height(5)
#        rectangle.set_width(10)
        Shapes_Cache.cache[rectangle.get_id()] = rectangle
  
# user

Shapes_Cache.load()

square = Shapes_Cache.get_shape("1")
print(square.get_type())

circle = Shapes_Cache.get_shape("2")
print(circle.get_type())

rectangle = Shapes_Cache.get_shape("3")
print(rectangle.get_type())
```
